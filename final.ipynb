```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv', sep=";", engine="python")
df_info = df.dtypes
df_head = df.head()

df_info
```

```python
df_head
```

```python
# Convert categorical variables to category type
categorical_columns = [
    'Marital status',
    'Application mode', 
    'Course',
    'Daytime/evening attendance\t',
    'Previous qualification',
    'Nacionality',
    "Mother's qualification",
    "Father's qualification", 
    "Mother's occupation",
    "Father's occupation",
    'Displaced',
    'Educational special needs',
    'Debtor', 
    'Tuition fees up to date',
    'Gender',
    'Scholarship holder',
    'International',
    'Target'
]

# Convert each column to category type
for col in categorical_columns:
    if col in df.columns:
        df[col] = df[col].astype('category')
```

```python
df_info
```

```python
df_head
```

```python
# Calculate pass rates for both semesters
df['sem1_pass_rate'] = df['Curricular units 1st sem (approved)'] / df['Curricular units 1st sem (enrolled)']
df['sem2_pass_rate'] = df['Curricular units 2nd sem (approved)'] / df['Curricular units 2nd sem (enrolled)']

# Drop original semester columns
df = df.drop(columns=[
    'Curricular units 1st sem (credited)',
    'Curricular units 1st sem (enrolled)', 
    'Curricular units 1st sem (evaluations)',
    'Curricular units 1st sem (approved)',
    'Curricular units 1st sem (grade)',
    'Curricular units 1st sem (without evaluations)',
    'Curricular units 2nd sem (credited)',
    'Curricular units 2nd sem (enrolled)',
    'Curricular units 2nd sem (evaluations)', 
    'Curricular units 2nd sem (approved)',
    'Curricular units 2nd sem (grade)',
    'Curricular units 2nd sem (without evaluations)'
])

# Replace NaN values with 0 in pass rate columns
df['sem1_pass_rate'] = df['sem1_pass_rate'].fillna(0)
df['sem2_pass_rate'] = df['sem2_pass_rate'].fillna(0)
```

```python
import statsmodels.api as sm

X = df.drop(columns=["Target"])  # 假设 "Target" 是因变量
Y = df["Target"]  # 目标变量

# 添加截距项（常数项）
X = sm.add_constant(X)
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

X_numeric = df[['Application order', 'Previous qualification (grade)', 'Admission grade', 'Age at enrollment', 'Unemployment rate', 'Inflation rate', 'GDP', 'sem2_pass_rate']]

X_numeric.head()
```

```python

X_with_const = sm.add_constant(X_numeric)

vif_data = pd.DataFrame()
vif_data["Feature"] = X_with_const.columns
vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]
vif_data
```

```python
target_mapping = {'Dropout': 0, 'Graduate': 2, 'Enrolled': 1}

Y = df['Target'].map(target_mapping)

X_categorical = df[categorical_columns]
X_categorical.drop('Target', axis=1, inplace=True)

X_categorical
```

```python
Y
```

```python
# Perform one-way ANOVA for each categorical variable
from scipy import stats

anova_results = {}
for col in categorical_columns:
    # Get unique categories
    categories = df[col].unique()
    
    # Create groups of target values for each category using the mapped target values
    groups = [Y[df[col] == cat].values for cat in categories]
    
    # Perform one-way ANOVA
    f_stat, p_val = stats.f_oneway(*groups)
    
    anova_results[col] = {
        'F-statistic': f_stat,
        'p-value': p_val
    }

# Display ANOVA results
# Convert ANOVA results to DataFrame
anova_df = pd.DataFrame.from_dict(anova_results, orient='index')
anova_df = anova_df.round(4)
anova_df = anova_df.sort_values('p-value')

anova_df
```

```python
# Drop features with high p-values from ANOVA test (p > 0.05)
features_to_drop = ['Nacionality', 'Educational special needs', 'International']
X_filtered = X_categorical.drop(features_to_drop, axis=1)
X_filtered
```

```python
# Use label encoding instead of one-hot encoding to reduce dimensionality
from sklearn.preprocessing import LabelEncoder

# Create label encoder
le = LabelEncoder()

# Apply label encoding to each categorical column
for col in X_filtered.columns:
    X_filtered[col] = le.fit_transform(X_filtered[col])

X_filtered

X_all = pd.concat([X_numeric, X_filtered], axis=1)
X_all
```

```python

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_filtered)

# 添加截距项
X_scaled = sm.add_constant(X_scaled)

# 重新训练 GLM
model = sm.GLM(Y, X_scaled, family=sm.families.Binomial())
result = model.fit()

# 显示结果
print(result.summary())
```

```python
model = sm.MNLogit(Y, sm.add_constant(X_scaled))
result = model.fit()

# 输出优化后的结果
print(result.summary())
```

```python
import statsmodels.api as sm
import pandas as pd
import numpy as np
from statsmodels.sandbox.regression.gmm import GMM


class MyGMM(GMM):
    def momcond(self, params):
        beta = params
        errors = self.endog - np.dot(self.exog, beta)
        return errors[:, None] * self.exog  # 计算矩条件

# 初始化 GMM (修正 instrument 参数)
gmm_model = MyGMM(Y, X_all, instrument=X_all)  # 传入 instrument 变量
params0 = np.zeros(X_all.shape[1])  # 初始参数值
gmm_results = gmm_model.fit(start_params=params0, optim_method='bfgs')

# 输出结果
print(gmm_results.summary())
```
